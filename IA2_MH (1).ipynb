# (1) Data Description 

## Description

The dataset was collected from four different databases: Cleveland, Hungary, Switzerland, and Long Beach V in 1988. The dataset has 76 attributes within the dataset. However, in all published experiments only **14 attributes** are referred to. Importantly, the target variable is the presence of heart disease. Heart disease is either 1 = present or 0 = absence implying a binary classification process can be used. There are 1025 observations within the dataset loaded with the 14 attributes.      

# Variables within the dataset 

## Categorical
* `Sex` (Gender of the patient (1 = male; 0 = females)
* `cp` (Chest pain type; 4 values = 0,1,2,3)
* `fbs` (Fasting blood sugar > 120 mg/dl (1 = true; 0 = false))
* `restecg` (Resting electrocardiographic results; 3 values = 0,1,2)
* `exang` (Exercise-induced angina; 2 values:1 = yes; 0 = no)
* `slope` (of the peak exercise ST segment; 3 values = 0,1,2)
* `ca` (Number of major vessels colored by fluoroscopy; 4 values = 0,1,2,3)
* `thal` (3 values: 1 = normal; 2 = fixed defect; 3 = reversible defect)
* `target` (1 or 0)

## Continuous
* `Age` (in years (of patient))
* `trestbps` (Resting blood pressure in mm Hg)
* `chol` (Serum cholesterol in mg/dl)
* `thalach` (Maximum heart rate achieved)
* `oldpeak` (ST depression induced by exercise relative to rest)

# (2) Question

Can we predict the presence of heart disease using binary classification by analyzing 4 key attributes (`thalach`, `oldpeak`, `chol`, `age`)?

**Response:** `target`

## Expansion on Question: 

To determine key attributes, we will aim to visualize through EDA 2-3 variables that are strongly associated with the presence of heart disease. Here, I will attempt to visualize the maximum heart rate achieved (`thalach`), ST depression induced by exercise relative to rest (`oldpeak`), `age`, and serum (`chol`)cholesterol (mg/dl). 

**Reasoning:** The reason I have chosen to use the three variables above is because of previous research surrounding these variables and their relationship to heart disease. It is well known, that `age` is a common reason for increased heart disease potential. For `thalach`, maximum heart rates indicate cardiovascular fitness level. Physical fitness has been previously noted to be beneficial in combatting heart disease for the unfit, but the relationship isn't as clear for the fit (Williams, 2010). Next, the degree of ST-wave depression induced by exercise relative to rest is possibly associated with a greater likelihood of heart disease as the peak reaches higher levels. Lastly, high serum levels have also been investigated to increase mortality in terms of heart disease as well (Jung, 2022). From these variables, we can investigate their relationship to heart disease through the use of **prediction**.

# Methods and Plan

## Proposed Method: 

I will use logistic regression with variables selected in this stage to address the previous research question: Can we predict the presence of heart disease using binary classification by analyzing selected key attributes? 

Variable selection will be repeated within this section to obtain a subset of variables for prediction. Variable selection will occur using Information Value (IV). Information Value (IV) will allow me to accurately understand the predictability of the variables within the dataset removing the ones with a weak relationship. Afterward, the logistic model will be built off the selected variables. Then to address multicollinearity, VIF will be conducted on the model. Lastly, interpretation will consider the original variable meaning and whether or not the results are consistent with real-life interpretation. 

### Assumptions and Issues

Selected variables will be used to predict `target`. This method was proposed after group discussion; along with another group member, I will be selecting a subset of variables to use to build a logistic regression model. This model will then be used to predict a binary response variable `target`. In this case, logistic regression was chosen due to the binary response of  either "0" (absent) or "1" (present) for heart disease in a patient. We are assuming linearity and independence in this model. One possible problem we might encounter in this model is multicollinearity. This could be an issue if our predictors are correlated with each other; in order to potentially account for this problem, I will be using a VIF to check the possible coefficient before analysis. 

# Implementation of a proposed model

```r
# Loading packages and dataset for IV, WOE

install.packages("Information")
library(Information)

heart_MH 

summary(heart_MH)

# Storing categorical variables as factors 

heart_MH$sex <- as.factor(heart_MH$sex)
heart_MH$cp <- as.factor(heart_MH$cp)
heart_MH$fbs <- as.factor(heart_MH$fbs)
heart_MH$restecg <- as.factor(heart_MH$restecg) 
heart_MH$exang <- as.factor(heart_MH$exang) 
heart_MH$slope <- as.factor(heart_MH$slope) 
heart_MH$ca <- as.factor(heart_MH$ca) 
heart_MH$thal <- as.factor(heart_MH$thal) 

```

```r
# Computing IV and WOE 

IV_heart <- create_infotables(data=heart_MH, y ="target", bins=10, parallel=FALSE)
IV_heart

IV_Value = data.frame(IV_heart$Summary)
```

```r
# Remove `fbs` 

heart_MH_IV <- heart_MH %>%
    select(-fbs)

### Model 


# Packages 
library(MASS)
library(tidyverse)
library(repr)
library(digest)
library(infer)
library(gridExtra)
library(mlbench)
library(AER)
library(ISLR)
library(broom)
library(qqplotr)
library(performance)
library(see)
library(glmbb)
library(cowplot)

# Logistic model using the 12 predictors selected above 
heart_model <- 
  glm(formula = target ~.,
      data = heart_MH_IV,
      family = binomial)

summary(heart_model)
```

```r
# Run VIF on remaining 12 variables 
install.packages("car")
library(car)

# Calculating VIF
vif_heart <- vif(heart_model)
vif_heart
```

```r
# Drop `slope`, `restecg`, `thal` <- Reasoning: Large p-vaules 

heart_MH_after <- heart_MH_IV %>%
    dplyr::select(age, sex, cp, trestbps, chol, thalach, exang, oldpeak, ca, target) 

heart_MH_after

# Recreate model with above predictors 
heart_model_2 <- 
  glm(formula = target ~.,
      data = heart_MH_after,
      family = binomial)

summary(heart_model_2)
```

```r
# Drop `age` <- Reasoning: High p-value // Refit model 

heart_MH_3 <- heart_MH_after %>%
    dplyr::select(sex, cp, trestbps, chol, thalach, exang, oldpeak, ca, target) 

heart_MH_3

# Recreate model with above predictors 
heart_model_3 <- 
  glm(formula = target ~.,
      data = heart_MH_3,
      family = binomial)


### FINAL MODEL AND RESULTS BELOW
summary(heart_model_3)
```

```r
# AUC 
install.packages("pROC")
library(pROC)

prediction <- predict(heart_model_3, heart_MH_3, 
                      type = "response")
roc_object <- roc(heart_MH_3$target, prediction)
auc(roc_object)
```
