# Methods and Plan

To address the question of interest—predicting the presence of heart disease using a subset of variables from the Heart Disease Dataset—I propose using **logistic regression** as the primary method of analysis.

### Why Logistic Regression?

Logistic regression is a widely used method for binary classification problems, where the response variable (target) has two possible outcomes—in this case, the presence (`1`) or absence (`0`) of heart disease. This method is particularly suitable for the following reasons:

- **Interpretable Results**: Logistic regression provides interpretable coefficients that quantify the relationship between the predictor variables and the odds of the response variable. For example, it will allow us to understand how age, cholesterol, or chest pain type impacts the likelihood of having heart disease.
- **Compatibility with Binary Target**: The target variable in this dataset is binary, making logistic regression a natural choice for modeling the probability of heart disease presence.
- **Flexibility**: Logistic regression can handle a mix of continuous variables (e.g., age, cholesterol, resting blood pressure) and categorical variables (e.g., sex, chest pain type, fasting blood sugar, and resting ECG results).

### Assumptions of Logistic Regression

The following assumptions are required to apply logistic regression:

1. **Independence of Observations**: Each observation in the dataset should be independent of others.
2. **Linear Relationship Between Predictors and Log-Odds**: The model assumes a linear relationship between the explanatory variables and the log-odds of the outcome.
3. **No Multicollinearity**: Predictor variables should not be highly correlated with one another. Multicollinearity can inflate the standard errors of the coefficients, reducing model reliability.
4. **Appropriate Distribution of Data**: While logistic regression does not require the predictors to be normally distributed, extremely skewed data may need transformation.

### Potential Limitations or Weaknesses

Despite its advantages, logistic regression has some limitations:

- **Linearity Assumption**: The method assumes a linear relationship between the predictors and the log-odds, which may not always hold true. If this assumption is violated, model performance can suffer.
- **Outliers**: Logistic regression is sensitive to extreme values in the data, which can disproportionately affect the model coefficients.
- **Overfitting with Many Predictors**: Including too many predictors relative to the sample size can lead to overfitting. This is mitigated here by pre-selecting a subset of variables.
- **Non-Complex Relationships**: Logistic regression is less effective in capturing complex, non-linear relationships between predictors and the response variable. Alternative methods like decision trees or random forests could be considered if non-linearity is suspected.

# Implementation of a Proposed Model

In this section, I implement a logistic regression model to predict the presence of heart disease (`target`) based on the selected variables (`age`, `chol`, `trestbps`, `sex`, `fbs`, `restecg`, and `cp`). The dataset is split into training and testing subsets to evaluate model performance. The results are summarized in a table showing the coefficients, p-values, and odds ratios of the predictors.

```r
set.seed(123)
train_index <- createDataPartition(heart_data$target, p = 0.8, list = FALSE)
train_data <- heart_data[train_index, ]
test_data <- heart_data[-train_index, ]

# Fit logistic regression model
log_model <- glm(target ~ age + trestbps + sex + restecg + cp, 
                 data = train_data, 
                 family = binomial)

# Summary of the model
model_summary <- summary(log_model)
print(model_summary)
```

```r
# Calculate odds ratios and confidence intervals
exp_coef <- exp(coef(log_model))  # Odds ratios
conf_int <- exp(confint(log_model))  # Confidence intervals
results_table <- cbind(Odds_Ratio = exp_coef, conf_int)

# Print the results table
print(results_table)
```

```r
# Predict on the test set and calculate accuracy
library(pROC)
test_data$predicted <- ifelse(predict(log_model, newdata = test_data, type = "response") > 0.5, 1, 0)
conf_matrix <- table(test_data$target, test_data$predicted)
accuracy <- sum(diag(conf_matrix)) / sum(conf_matrix)
roc_curve <- roc(test_data$target, predict(log_model, newdata = test_data, type = "response"))
cat("Model Accuracy: ", accuracy, "\n")
```

```r
auc_value <- auc(roc_curve)
cat("AUC: ", auc_value, "\n")
```

### Visualization of the results

```r
# Ensure target variable is a factor with correct levels
test_data$target <- factor(test_data$target, levels = c(0, 1))
plot(roc_curve, main = "ROC Curve for Logistic Regression Model", col = "blue", lwd = 2)
```

- The logistic regression model shows good predictive power with an accuracy of 77.94% and an AUC of 0.828, indicating a strong ability to distinguish between patients with and without heart disease.
- Significant predictors include age, cholesterol levels, resting blood pressure, sex, chest pain types, and resting electrocardiogram results, all of which are known risk factors for heart disease.
