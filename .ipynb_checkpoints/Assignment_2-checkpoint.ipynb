- ```target```: presence of heart disease in the patient (0 = no disease, 1 = disease), binary categorical variable

### Helpful definitions:
- **Angina**: chest pain resulting from the heart not receiving enough oxygen-rich blood
- **ST Segment**: The ST segment appears in the ECG results. It represents the period of the cardiac cycle where the muscular layer of the heart contracts to expel blood from the ventricles (lower chambers of the heart).
- **ST Depression**: If there is abnormally low activity in the ST segment, the person is said to have ST depression.
- **Thalassemia**: an inherited blood disorder caused when the body doesn't make enough hemoglobin

Forward and backward selection are both stepwise feature selection methods designed to produce the most accurate model given a list of variables. Given that the goal is to produce a model that can accurately and precisely predict the presence of heart disease, it seems fitting to use algorithms that are designed to produce the most accurate possible model. While there is debate over whether or not stepwise selection algorithms such as forward and backward selection are valid methods (Flom & Cassell, 2007), I believe that the models they produce will be quite accurate. **Which algorithm (forward/backward selection, LASSO, Ridge regression) produces the most accurate/precise model?**

No assumptions are required to apply forward or backwards selection. The most plausible potential limitation is that the algorithm may not select the best possible model. Stepwise selection methods have been labelled as controversial due to their tendenceies to bias standard errors and p-values towards zero and bias parameter estimates away from zero (Flom & Cassell, 2007). This can cause the outputted models to be less accurate, and possibly not include the most significant variables (Flom & Cassell, 2007). Due to this, the model derived from these algorithms may not be as accurate as possible. This is why I propose to compute a confusion matrix for both models. ## 2. Implementation of a Proposed Method

````r
library(repr)
library(digest)
library(infer)
library(gridExtra)
library(faraway)
library(broom)
library(leaps)
library(mltools)
library(tidyverse)
library(caret)
library(pROC)
library(boot)
library(glmnet)
````

````r
# splitting data into testing and training set
data <- data %>%
    mutate(ID = row_number())

training_data <- 
    data %>%
    slice_sample(prop = 0.70, replace = FALSE)

selection_data <- 
    data %>%
    anti_join(training_data, by = "ID")

training_data <- 
    training_data %>% 
    select(-"ID")

selection_data <- 
    selection_data %>% 
    select(-"ID")

head(training_data)
nrow(training_data)

head(selection_data)
nrow(selection_data)
````

````r
# using backwards selection
backward_sel <- 
    regsubsets(
        x = target ~ ., 
        nvmax = 22,
        data = selection_data,
        method = "backward",
    )

backward_sel_summary <- summary(backward_sel)

backward_sel_summary_df <- tibble(
    n_input_variables = 1:22,
    RSQ = backward_sel_summary$rsq,
    RSS = backward_sel_summary$rss,
    ADJ.R2 = backward_sel_summary$adjr2
)

backward_sel_summary
backward_sel_summary_df
````

````r
# building model based on results of backward selection
backward_sel_model <- glm(target ~ sex + cp + trestbps + chol + restecg + thalach + oldpeak + slope + ca + thal, 
                           data = training_data, 
                           family = "binomial")
````

````r
# building the confusion matrix
backward_pred_class <- 
  round(predict(backward_sel_model, type = "response"), 0)

backward_confusion_matrix <- 
    confusionMatrix(
    data = as.factor(backward_pred_class),
    reference = as.factor(training_data$target),
    positive = '1'
)
````

````r
#same thing, this time for forward selection
foreward_sel <- 
    regsubsets(
        x = target ~ ., 
        nvmax = 22,
        data = selection_data,
        method = "forward",
    )

foreward_sel_summary <- summary(foreward_sel)

foreward_sel_summary_df <- tibble(
    n_input_variables = 1:22,
    RSQ = forward_sel_summary$rsq,
    RSS = forward_sel_summary$rss,
    ADJ.R2 = forward_sel_summary$adjr2
)

foreward_sel_summary
foreward_sel_summary_df

foreward_sel_model <- glm(target ~ age + sex + cp + trestbps + restecg + thalach + exang + oldpeak + slope + ca + thal, 
                           data = training_data, 
                           family = "binomial")

foreward_pred_class <- 
  round(predict(foreward_sel_model, type = "response"), 0)

foreward_confusion_matrix <- 
    confusionMatrix(
    data = as.factor(foreward_pred_class),
    reference = as.factor(training_data$target),
    positive = '1'
)
````

````r
#comparing confusion matrices
backward_confusion_matrix
foreward_confusion_matrix

#we can see that despite the minor differences in model selection, there is little difference in prediction performance
````

````r
# ROC curve for backward selection model
ROC_backward <- 
    roc(response = training_data$target, 
        predictor = predict(backward_sel_model, type = "response"))

plot(ROC_backward,
  print.auc = TRUE, col = "blue", lwd = 3, lty = 2,
  main = "ROC Curves for Heart Disease Dataset (Backward Selection)")
````
